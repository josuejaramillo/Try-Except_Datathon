# Indice de gentrificacion


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#abrir csv Datos_del_hogar.csv
datos_hogar = pd.read_csv('Datos_del_hogar.csv',encoding='latin1',sep=';')
antioquia= datos_hogar[datos_hogar["DPTO"]==5]
data = pd.DataFrame()
data["arriendo"] = antioquia["P5140"]
data["pagaria_arriendo"] = antioquia["P5130"]
data["precio_venta"]= antioquia["P5110"]
data["area"]= antioquia["AREA"]
data["directorio"]= antioquia["DIRECTORIO"]
#histograma de arriendo
# data[data["arriendo"] != "NaN"].hist(column="arriendo",bins=50)
```

```python
data[data["area"]==5].hist(column="arriendo",bins=1000)
plt.xlim(0, 2000000)
```

```python
# Filtrar los datos donde el área es 5
data_area_5 = data[data["area"] == 5]

# Calcular el percentil 25 del arriendo
percentil_25_arriendo = data_area_5["arriendo"].quantile(0.25)

print("Percentil 25 de arriendo:", percentil_25_arriendo)
```
